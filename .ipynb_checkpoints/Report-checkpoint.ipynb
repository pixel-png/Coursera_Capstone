# Introduction

I'm thinking of moving to a new town in Tokyo Japan.​

The following four conditions are required for the destination.​

1. Be able to reach the company in less than 30 minutes by motorcycle.​

2. An area with few motorcycle theft cases.​

3. Close to the station​

4. The characteristics of the town are similar to Nakano Ward in Tokyo Japan where I live now​

# Data

1. Station Datasets in Tokyo.

    Station Name, location information



2. Motorcycle theft cases Datase

    Each motorcyle theft case information. Date and time information and location information.​


3. Foursquare venue information

   I used this data to compare the charasteristic with nakano ward.​


4. Dataset of  Time required from each station to the company

    This dataset is obtained from google map api.​

# Methodology

1. Get list of all stations in Tokyo
2. Calculate the time required from each station to the company using Google Map API
3. Narrow the previous station list to those that will arrive at the company within 30 minutes.
4. Get the Foursquare venue information of each town in Tokyo
5. Classify each town into 5 groups using k-means clustering
6. Narrow the previous station list to those that is exist in the town which is in the same group with where i live now.
7. Get the motorcycle theft risk information of each town.
8. Narrow the station list to stations with low risk of theft.

Finaly, I get the stations list where I should live in.
